```python
from os.path import splitext, basename
from rubin_sim import maf

from rubin_sim.data import get_data_dir
from rubin_sim.data import get_baseline

%matplotlib inline
```

```python
opsim_fname = get_baseline()
run_name = splitext(basename(opsim_fname))[0]
```

```python
constraint = "night < 61 and filter='r'"
slicer = maf.HealpixSlicer(nside=64)
metric = maf.Coaddm5Metric()
bundle = maf.MetricBundle(metric,slicer,constraint)
bg = maf.MetricBundleGroup([bundle],opsim_fname)
bg.run_all()
bg.plot_all(closefigs=False)
```
