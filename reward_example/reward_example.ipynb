```python
import healpy as hp
import matplotlib.pylab as plt
%matplotlib inline

from fbs_config_sv_survey import get_scheduler

from rubin_scheduler.scheduler.model_observatory import ModelObservatory
from rubin_scheduler.scheduler import sim_runner
```

```python
nside, scheduler = get_scheduler()
```

```python
# Can peak inside at the footprints
scheduler.survey_lists[3][2].basis_functions[2].footprint.footprints
```

```python
for bandname in 'ugrizy':
    hp.mollview(scheduler.survey_lists[3][2].basis_functions[2].footprint.get_footprint(bandname), title=bandname)
```

```python
mjd_start = 60857
observatory = ModelObservatory(nside=nside, mjd_start=mjd_start)
```

```python
sim_duration=10
observatory, scheduler, observations = sim_runner(
        observatory,
        scheduler,
        sim_duration=sim_duration,
        filename=None,
        verbose=True,
    )
```

```python
conditions = observatory.return_conditions()
```

```python
scheduler.survey_lists[4]
```

```python
for survey in scheduler.survey_lists[4]:
    hp.mollview(survey.calc_reward_function(conditions), title=survey.survey_name, min=0, max=11)
```

```python
# Jump ahead in time ~ 0.5 hours
observatory.mjd += .02
```

```python
later_conditions = observatory.return_conditions()
```

```python
# Now what do the reward functions look like now?
for survey in scheduler.survey_lists[4]:
    hp.mollview(survey.calc_reward_function(later_conditions), title=survey.survey_name, min=0, max=11)
```

```python
# Now the y+y is the winner! The high reward pixels in r+i and i+z from before got masked off.

# So a small change in time can result in a very different decision on what to observe by the scheduler.
```
