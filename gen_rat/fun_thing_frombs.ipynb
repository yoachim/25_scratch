```python
import numpy as np
from scipy.stats import binned_statistic
n_max = 8
n_fam = 10000000
input_rat = 0.5122
np.random.seed(42)
baby_array = np.random.rand(n_fam, n_max)
b_indx = np.where(baby_array <= input_rat)
first_b_indx, _be, _bn = binned_statistic(b_indx[0], b_indx[1],
                                          statistic=np.min,
                                          bins=np.arange(n_fam+1)-0.5)
no_boys = ~np.isfinite(first_b_indx) 
num_g = np.nansum(first_b_indx) + n_max*np.sum(no_boys)
num_b = n_fam - np.sum(no_boys)
print("output fraction of boys", num_b/(num_g+num_b))
```
