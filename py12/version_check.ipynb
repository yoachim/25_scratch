```python
import pandas as pd
```

```python
m2 = pd.read_csv('m2_p12.txt', delim_whitespace=True, comment='#', header=None)
usdf = pd.read_csv('usdf_py12.txt', delim_whitespace=True, comment='#', header=None)
```

```python
m2_dict = {}
for key,val in zip(m2[0],m2[1]):
    m2_dict[key] = val

usdf_dict = {}
for key,val in zip(usdf[0],usdf[1]):
    usdf_dict[key] = val
```

```python
for key in m2_dict:
    if key in usdf_dict.keys():
        if m2_dict[key] != usdf_dict[key]:
            print(key, m2_dict[key], usdf_dict[key])
```
